# Demonstrating how to work with Galaxy Zoo 2 Data!

```python
#First: Import some packages
from astropy.table import Table #for reading tables
import matplotlib.pyplot as plt # for making plots
%matplotlib inline # for displaying plots - only needed in a notebook
```

# 1: Reading the data and viewing galaxy images

```python
#read data 
#contains vote fractions from Galaxy Zoo and metadata from SDSS for 243,500 galaxies
data = Table.read('GZ2data.fits')
```

```python
#displays first row of table - python indexing starts at 0. data[1] would be the second row. 
data[0]
```

```python
#print list of RA, DEC for first 10 galaxies:
for i in range(0,10):
    print data[i]['RA'], data[i]['DEC']
```

## Next: we'll look at images of galaxies with certain properties. First, let's try looking at only spiral galaxies. These tend to have large vote f_features vote fractions. We'll get a list of ra and dec's of galaxies with f_features > 0.8 to start. 

```python
#From the data[0] cell above, we can copy/paste the column name that corresponds to f_features:
f_features = 't01_smooth_or_features_a02_features_or_disk_debiased' #this is the fraction of people who voted 'features or disk' to the first question in GZ2.
```

```python
#define spirals as galaxies in 'data' with f_features > 0.8:
spirals = (data[f_features]>0.8)
```

```python
#now if we put [spirals] next to data each time we select stuff, it will only consider rows that meet that criterion

#let's get another ra and dec list, but this time only for galaxies that are probably spirals

for i in range(0,10):
    print data[spirals][i]['RA'], data[spirals][i]['DEC']
```

# 2: Check distribution of galaxy properties in the sample

```python
#Let's see what redshift range our data is in:
plt.rcParams['xtick.labelsize'] = 15 #set size of tick numbers in plot (they're kind of small by default)
plt.rcParams['ytick.labelsize'] = 15 


f = plt.figure(figsize=(8,6)) #set figure size so it's easy to see. Feel free to edit/play
plt.hist(data['REDSHIFT']) # plot a histogram of the redshifts
plt.xlabel('redshift',fontsize=20) # label x axis
plt.ylabel('count',fontsize=20) # label y axis
plt.title('redshift distribution of GZ2 galaxies',fontsize=20) #title if you want
```

## So the above plot shows that the galaxies have redshifts out at 0.25, but most are around 0.05.
